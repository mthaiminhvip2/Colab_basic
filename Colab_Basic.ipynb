```python
import pandas as pd
import numpy as np

x = np.array([30,40,50,60,70,80,90,100,110,120])
y = np.array([145,213,240,320,340,420,470,510,560,600])

n = len(x)

w = 0
b = 0

alpha = 0.0001

echos = 10

for echo in range(echos):
  y_pred = x*w+b
  error = y_pred - y

  J = (1/n)*np.sum(error**2)

  dw = (2/n)*np.sum(error*x)
  db = (2/n)*np.sum(error)

  w = w - alpha*dw
  b = b - alpha*db

  print(f"hàm mất mát sau {echo} lần huấn luyện là: {J}")
  print(f"trọng số w và b sau {echo} lần huấn luyện là: {w} và {b}")
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/Colab_Basic/housing2.csv')
df.head()

n = len(df)

# 1xn
x1 = df['housing_median_age'].to_numpy().reshape(1,n)
x2 = df['total_rooms'].to_numpy().reshape(1,n)
x3 = df['total_bedrooms'].to_numpy().reshape(1,n)
y = df['median_house_value'].to_numpy().reshape(1,n)


# nx3
X = df[['housing_median_age' , 'total_rooms' , 'total_bedrooms']].to_numpy()

# 3xn
X = X.T

w0 = 0
w1 = 0
w2 = 0
w3 = 0

alpha = 0.001
echos = 10

for echo in range(echos):
    # 1x3
    w = np.array([w1,w2,w3]).reshape(1,3)
    print(f"w: {w}")

    # 1xn
    wbs = np.full((1,n) , w0)
    print(f"wbs: {wbs}")
    # 1xn
    Loss_function = np.dot(w , X) + wbs - y
    print(f"Loss_function : {Loss_function}" )

    loss_value = (1/n)*np.sum(Loss_function**2)
    dw0 = (2/n)*np.sum(Loss_function)
    dw1 = (2/n)*np.dot(Loss_function , x1.T).item()
    dw2 = (2/n)*np.dot(Loss_function , x2.T).item()
    dw3 = (2/n)*np.dot(Loss_function , x3.T).item()

    w1 = w1 - alpha*dw1
    w2 = w2 - alpha*dw2
    w3 = w3 - alpha*dw3
    w0 = w0 - alpha*dw0
    print(f"hàm mất mát sau {echo} lần huấn luyện là: {loss_value}")
    print(f"trọng số w0,w1,w2,w3 sau {echo} lần huấn luyện là: {w1} và {w2} và {w3} và {w0}")






```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Đọc dữ liệu
df = pd.read_csv('/content/drive/MyDrive/Colab_Basic/Housing.csv')  # Đổi đường dẫn nếu cần

# 2. Tách input (X) và output (y)
X = df[['housing_median_age', 'total_rooms', 'total_bedrooms']].to_numpy()   # shape: (n, 3)
y = df['median_house_value'].to_numpy()                           # shape: (n,)

# 3. Chuẩn hóa dữ liệu (feature scaling)
X_mean = np.mean(X, axis=0)
X_std  = np.std(X, axis=0)
X_norm = (X - X_mean) / X_std  # Chuẩn hóa theo từng cột

# 4. Thêm cột 1 vào đầu để tạo hệ số w0
n = X_norm.shape[0]
X_b = np.c_[np.ones((n, 1)), X_norm]  # shape: (n, 4)

# 5. Khởi tạo trọng số và tham số
w = np.zeros(X_b.shape[1])  # [w0, w1, w2, w3]
learning_rate = 0.01
epochs = 1000

# 6. Hàm mất mát
def compute_cost(X, y, w):
    n = len(y)
    y_pred = X @ w
    error = y_pred - y
    cost = (1 / (2 * n)) * np.dot(error, error)
    return cost

# 7. Gradient Descent
costs = []
for epoch in range(epochs):
    y_pred = X_b @ w
    error = y_pred - y
    gradients = (1 / n) * X_b.T @ error      #(4,n) . (n,) = (4,)
    w = w - learning_rate * gradients
    cost = compute_cost(X_b, y, w)
    costs.append(cost)
    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Cost: {cost:.2f}")

# 8. Dự đoán
def predict(input_data):  # input_data là [age, bedrooms, kitchens]
    input_norm = (np.array(input_data) - X_mean) / X_std
    input_vec = np.insert(input_norm, 0, 1.0)  # Thêm hệ số w0
    return input_vec @ w

# Ví dụ dự đoán: nhà 8 tuổi, 2 phòng ngủ, 1 bếp
print("Dự đoán giá:", predict([8, 2, 1]))

# 9. Vẽ đồ thị hàm mất mát
plt.plot(costs)
plt.xlabel('Epoch')
plt.ylabel('Cost (MSE)')
plt.title('Gradient Descent Loss')
plt.grid(True)
plt.show()
```
